```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ship = sns.load_dataset('titanic')
ship
```

```python
ship.isnull().sum()
```

```python
ship.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',hue= 'sex', data=ship)
```

```python
plt.scatter(x='age',y= 'fare', data=ship) # this comes from matplotlib library
```

```python
sns.scatterplot(x='age',hue= 'pclass',y= 'fare' , data=ship) # this comes from seaborn library
```

```python
sns.scatterplot(x='age',hue= 'sex',y= 'fare' , data=ship)
```

```python
sns.barplot(data=ship , x='sex',y= 'fare' )
```

```python
sns.pairplot(ship)
```

```python
sns.pairplot(ship[['age']]) 
```

```python
sns.histplot(ship)
```

```python
ship.hist(bins=10,figsize=(8,9))
```

```python
ship.hist(figsize=(8,9),grid=False)
```

```python
sns.boxplot(data=ship , x='sex',y= 'fare',hue= 'parch')
```

```python
sns.boxplot(data=ship , x='sex',y= 'fare',hue= 'pclass')
```

```python
sns.boxenplot(data=ship , x='sex',y= 'fare',hue= 'pclass')
```

```python
sns.boxenplot(data=ship , x='sex',y= 'fare',hue= 'parch')
```

```python
gridwise = sns.FacetGrid(ship, hue='survived', col='pclass',margin_titles=True,palette={1:'seagreen',0:'gray'})
gridwise = gridwise.map(plt.scatter, 'fare','age', edgecolor='w').add_legend()
```

```python
gridwise = sns.FacetGrid(ship, hue='survived', col='alone',margin_titles=True,palette={1:'seagreen',0:'gray'})
gridwise = gridwise.map(plt.scatter, 'fare','age', edgecolor='w').add_legend()
```
